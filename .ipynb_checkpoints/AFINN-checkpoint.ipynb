```python
start=-22.239634
stop=22.239634
step=1

df['pred']=df['afinn_adjusted'].apply(lambda x :"truthful" if x >= -13 and x<= 31 else "deceptive" )
df
```

```python
truePositive = trueNegative = falsePositive = falseNegative = 0
```

```python
for i in range(len(df)):
    if(df['deceptive'][i] == df['pred'][i] == "truthful"):
        truePositive += 1
    elif df['deceptive'][i] == df['pred'][i] == "deceptive":
        trueNegative += 1
    elif df['deceptive'][i] == "deceptive" and df['pred'][i] == "truthful":
        falsePositive += 1
    elif df['deceptive'][i] == "truthful" and df['pred'][i] == "deceptive" :
        falseNegative += 1
        
falseNegative
```

```python
accuracy = (truePositive + trueNegative)/(truePositive + trueNegative + falsePositive + falseNegative)
precision = truePositive/(truePositive + falsePositive)
recall =truePositive/(truePositive + falseNegative) 

accuracy,precision,recall
```
